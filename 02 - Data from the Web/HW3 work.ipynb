```python
def createDFfromURL(url, payload):
    req = requests.get(url, params=payload)
    data = bs(req.content, "html.parser")
    dfs = []
    for elem in data.find_all("tr")[1].next_siblings:
        df = pd.DataFrame(elem.strings, index=elem.t)
        df = df.transpose()
        dfs.append(df)
    mainDF = pd.concat(dfs, axis=0)
    mainDF.columns = ["Title", "Name", "Here", "Sciper"]
    mainDF.set_index("Sciper", inplace=True)
    return mainDF
```
