Using Postman, it is possible to find out the fields that interest us with the wanted values, that are shown below:

```python
url = "http://isa.epfl.ch/imoniteur_ISAP/!GEDPUBLICREPORTS.html?ww_x_GPS=1378362092&ww_i_reportModel=133685247&ww_i_reportModelXsl=133685270&ww_x_UNITE_ACAD=249847&ww_x_PERIODE_ACAD=123456101&ww_x_PERIODE_PEDAGO=null&ww_x_HIVERETE=null"
```

```python
r = requests.get(url)
```

```python
df = pd.read_html(r.content, header = 1)[0]
```
